```python
import matplotlib.pyplot as plt
import numpy as np
X = np.arange(-6,6,0.0002)

def relu(X):
    y = [1 if (x >= 0) else 0 for x in X]
    y = np.array(y)
    return y


def logistic_sigmoid(X):
    fx = 1/(1 + np.exp(-X))
    y = fx*(1 - fx)
    return y

    
def piece_wise_linear(X):
    y = [0.1 if (x>1) else 1 if (x >= -1 and x <= 1) else 0.1 for x in X]
    y = np.array(y)
    return y

def swish_fun(a,X):
    fx = 1/(1 + np.exp(-a*X))
    y = fx + a*X*fx*(1-fx)
    return y

def exponential_linear(a,X):
    y = [1 if (x > 0) else a*(np.exp(x)) for x in X]
    y = np.array(y)
    return y
   


def plot_function(y,X,title):
    fast_learning_region = np.where(y > 0.9)[0]
    active_learning_region = np.where((y >= 0.05) & (y <= 0.9))[0]
    slow_learning_region = np.where(y < 0.05)[0]
    inactive_learning_region = np.where(y == 0.0)[0]
    plt.plot(X[slow_learning_region], y[slow_learning_region],color='y',label='Slow Learning')
    plt.plot(X[active_learning_region], y[active_learning_region], color='g',label='Active Learning')
    plt.plot(X[fast_learning_region], y[fast_learning_region],color='b',label='Fast Learning')
    plt.plot(X[inactive_learning_region], y[inactive_learning_region],color='m',label='Inactive Learning')
    plt.ylabel("f'(z)")
    plt.xlabel("z")
    plt.title(title)
    plt.legend(loc='upper left') 
    plt.show()



y = relu(X)
plot_function(y,X,'Rectified linear unit')


y = logistic_sigmoid(X)
plot_function(y,X,'Logistic sigmoid activation function')

y = piece_wise_linear(X)
plot_function(y,X,'Piece-wise linear unit')

y = swish_fun(4,X)
plot_function(y,X,'Swish')
  
y = exponential_linear(0.25,X)  
plot_function(y,X,'Exponential Linear unit')


```
